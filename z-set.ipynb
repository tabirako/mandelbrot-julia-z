```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#change these parameters
res_x,res_y = 640,480
center_x, center_y = 1.05,1.01
zoom = 100
z0 = 0 + 0j
c0 = 0.5 + -0.5*1j


zoom = 1/zoom
start_x, end_x = center_x - zoom, center_x + zoom
start_y, end_y = center_y + zoom, center_y - zoom
```

```python
real_points = np.linspace(start_x,end_x,num=res_x)
imag_points = np.linspace(start_y,end_y,num=res_y)
```

```python
real_grid, imag_grid = np.meshgrid(real_points, imag_points)
```

```python
complex_array = real_grid + imag_grid * 1j
```

```python
complex_color = np.zeros((res_y,res_x))
```

```python
steps = 250
thre = 10**4

def abs_size(p):
        return p.real**2 + p.imag**2
    
def color(p):
    z = z0
    for i in range(steps):
        z = z**p + c0
        if abs_size(z) >= thre:
         return i
    return steps*2
    
for row in range(res_y):
    for col in range(res_x):
        complex_color[row,col] = color(complex_array[row,col])
```

```python
plt.imshow(complex_color,norm='asinh', interpolation='none',cmap='gist_ncar_r', extent=[start_x,end_x,end_y,start_y])
plt.show()
```
